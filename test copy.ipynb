```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Apply the custom style
plt.style.use('./dharmstyledarkmode.mplstyle')

# Set up some sample data
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2022 = np.random.normal(100000, 15000, 12)
sales_2023 = sales_2022 * np.random.normal(1.1, 0.1, 12)
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
category_sales = np.random.normal(50000, 10000, 5)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(months, sales_2022, marker='o', linewidth=2.5, label='2022 Sales')
plt.plot(months, sales_2023, marker='s', linewidth=2.5, label='2023 Sales')

plt.title('Monthly Sales Comparison (2022 vs 2023)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales ($)', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)

# Format y-axis to show currency
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, category_sales, alpha=0.8)

plt.title('Sales by Product Category', fontsize=16, fontweight='bold')
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)

# Add value labels on bars
for bar, value in zip(bars, category_sales):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1000, 
             f'${value:,.0f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
# Generate correlated data
marketing_spend = np.random.normal(5000, 1500, 50)
sales_revenue = marketing_spend * 8 + np.random.normal(0, 5000, 50)
store_sizes = np.random.choice(['Small', 'Medium', 'Large'], 50)

plt.figure(figsize=(10, 6))

# Create scatter plot with different colors for store sizes
for i, size in enumerate(['Small', 'Medium', 'Large']):
    mask = store_sizes == size
    plt.scatter(marketing_spend[mask], sales_revenue[mask], 
               alpha=0.7, s=80, label=f'{size} Stores')

plt.title('Sales Revenue vs Marketing Spend by Store Size', fontsize=16, fontweight='bold')
plt.xlabel('Marketing Spend ($)', fontsize=14)
plt.ylabel('Sales Revenue ($)', fontsize=14)
plt.legend()

# Add trend line
z = np.polyfit(marketing_spend, sales_revenue, 1)
p = np.poly1d(z)
plt.plot(sorted(marketing_spend), p(sorted(marketing_spend)), "--", alpha=0.8, linewidth=2)

plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
# Generate sales distribution data
daily_sales = np.random.gamma(2, 15000, 1000)

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(daily_sales, bins=30, alpha=0.7, edgecolor='white', linewidth=0.5)

plt.title('Distribution of Daily Sales', fontsize=16, fontweight='bold')
plt.xlabel('Daily Sales ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Add vertical line for mean
mean_sales = np.mean(daily_sales)
plt.axvline(mean_sales, color='red', linestyle='--', linewidth=2, alpha=0.8, 
           label=f'Mean: ${mean_sales:,.0f}')

plt.legend()
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

plt.tight_layout()
plt.show()
```
